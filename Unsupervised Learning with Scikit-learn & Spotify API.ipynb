# Building a Table

```python
import pandas as pd
 
names = []
followers = []
popularity = []
genres = []
urls = []
 
for i in json_results:
  names.append(i['name'])
  followers.append(i['followers']['total'])
  popularity.append(i['popularity'])
  genres.append(i['genres'])
  urls.append(i['external_urls']['spotify'])
 
df = pd.DataFrame()
df['names'] = names
df['followers'] = followers
df['popularity'] = popularity
df['genre'] = genres
df['url'] = urls
 
df.head()
```

```python
by_followers = df[['names','followers']].sort_values(by='followers', ascending=False)
by_followers.head(10)
```

```python
by_popularity = df[['names','popularity']].sort_values(by='popularity', ascending=False)
by_popularity.head(10)
```

# Clustering with KMeans

```python
from sklearn.cluster import KMeans
import numpy as np
X = df[['followers','popularity']]
 
from sklearn.preprocessing import MinMaxScaler
s = MinMaxScaler()
Scaled_X = s.fit_transform(X)
 
kmeans = KMeans(n_clusters=8, random_state=0).fit(X)
y_kmeans = kmeans.predict(X)
 
Scaled_Kmeans = KMeans(n_clusters=8, random_state=0).fit(Scaled_X)
Sy_kmeans = Scaled_Kmeans.predict(Scaled_X)
 
df['labels'] = kmeans.labels_
df['scaled_labels'] = Scaled_Kmeans.labels_
 
plt.scatter(X.followers, X.popularity, c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
final = df[['names','followers','popularity','labels','scaled_labels']].sort_values(by='labels', ascending=False)
display(final)
final.to_excel('/content/drive/My Drive/Houston_Artists_Categories_8-7-2020.xlsx')
```
